# 1. areaWithType


```python
import re
```

```python
def get_builtup_area(text):
    match = re.search(r'Super Built up area (\d+\.?\d*)',text)
    if match:
        return float(match.group(1))#group(1) is the first parenthesized subgroup in the regex
    return None
```

```python
def get_area(text,area_type):
    match = re.search(area_type +r'\s*:\s*(\d+\.?\d*)',text)
    if match:
        return float(match.group(1))
    return None
```

```python
def convert_to_sqrt(text,area_value):
    if area_value is None:
        return None
    match = re.search(r'{} \((\d+\.?\d*) sq.m.\)'.format(area_value),text)
    if match:
        sq_m_value = float(match.group(1))
        return sq_m_value * 10.7639
    return area_value
```

```python
df['super_builtup_area'] = df['areaWithType'].apply(get_builtup_area)
df['super_builtup_area'] = df.apply(lambda row: convert_to_sqrt(row['areaWithType'],row['super_builtup_area']), axis=1)

df['carpet_area'] = df['areaWithType'].apply(lambda x: get_area(x,'Carpet area'))
df['carpet_area'] = df.apply(lambda row: convert_to_sqrt(row['areaWithType'],row['carpet_area']), axis=1)


df['built_up_area'] = df['areaWithType'].apply(lambda x: get_area(x,'Built Up area'))
df['built_up_area'] = df.apply(lambda row: convert_to_sqrt(row['areaWithType'],row['built_up_area']), axis=1)
```

# 2. Additional room

```python
def convert_scale(row):
    if np.isnan(row['area']) or np.isnan(row['built_up_area']):
        return row['built_up_area']
    else:
        if round(row['area']/row['built_up_area']) == 9.0:
            return row['built_up_area'] * 9
        elif round(row['area']/row['built_up_area']) == 11.0:
            return row['built_up_area'] * 10.7
        else:
            return row['built_up_area']
```

```python
df_nan_df['built_up_area'] = df_nan_df.apply(convert_scale, axis=1)
```

# 3. age_possession

```python
def categorize_age_possession(value):
    if pd.isna(value):
        return "Undefined"
    
    value = str(value)

    if "0 to 1 Year Old" in value or "Within 6 months" in value or "within 3 months" in value:
        return "New Property"
    
    if "1 to 5 Year Old" in value:
        return "Relatively New"
    
    if "5 to 10 Year Old" in value:
        return "Moderately Old"
    
    if "10+ Year Old" in value:
        return "Old Property"
    
    if "Under Construction" in value or "By" in value:
        return "Under Construction"
    
    try:
        # For entries like 'May 2024'
        int(value.split(" ")[-1])
        return "Under Construction"
    except:
        return "Undefined"


```

```python
# Apply to dataframe column
df["agePossession"] = df["agePossession"].apply(categorize_age_possession)
```

```python
df['furnishDetails']
```

```python
import ast
import re

# Convert string representation of list into actual list
df["furnishDetails"] = df["furnishDetails"].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x 
    # ast.literal_eval safely evaluates a string containing a Python literal (like a list) into the corresponding Python object
)
```

```python
all_furnishings = []

for details in df["furnishDetails"].dropna():
    for item in details:
        # Remove numbers and "No"
        clean_item = re.sub(r"No\s+|\d+\s*", "", item).strip()
        if clean_item:
            all_furnishings.append(clean_item)

columns_to_include = list(set(all_furnishings))
```

```python
def get_furnishing_count(details, furnishing):
    if isinstance(details, list):
        for item in details:
            if f"No {furnishing}" in item:
                return 0
            
            match = re.search(rf"(\d+)\s*{re.escape(furnishing)}", item)
            if match:
                return int(match.group(1))
            
            if furnishing == item.strip():
                return 1
    return 0
```

```python
for furnishing in columns_to_include:
    df[furnishing] = df["furnishDetails"].apply(
        lambda x: get_furnishing_count(x, furnishing)
    )

furnishings_df = df[["furnishDetails"] + columns_to_include]
```

```python
furnishings_df.drop(columns=["furnishDetails"], inplace=True)
```

```python
furnishings_df.sample(10)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(furnishings_df)
```

```python
wcss_reduced =[]

for i in range(1,11):
    kmenas = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmenas.fit(scaled_data)
    wcss_reduced.append(kmenas.inertia_)
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,11), wcss_reduced, marker='o',linestyle='--')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
clusters = 3
#fit the model
kmeans = KMeans(n_clusters=clusters, random_state=42)
kmeans.fit(scaled_data)

#predicted clusters assignment for each row
cluster_assignments = kmeans.predict(scaled_data)
```

```python
df['furnishing_type'] = cluster_assignments
```
