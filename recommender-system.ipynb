```python
tfidf_vec = TfidfVectorizer(stop_words='english',ngram_range=(1,2))
```

```python
tfidf_matrix = tfidf_vec.fit_transform(df['Facilitiesstr'])
```

```python
def recommend_appartments(title, cosine_sim1=cosine_sim1):
    # Get the index of the apartment that matches the title
    idx = df.index[df['PropertyName'] == title][0]
    # Get the pairwise similarity scores of all apartments with that apartment
    sim_scores = list(enumerate(cosine_sim1[idx]))
    # Sort the apartments based on the similarity scores
    # sim_scores is a list of tuples where the first element is the index of the apartment 
    # and the second element is the similarity score. We sort this list in descending order
    # based on the similarity score.
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get the scores of the 5 most similar apartments
    sim_scores = sim_scores[1:6]
    # Get the apartment indices
    apartment_indices = [i[0] for i in sim_scores]

    recommendations = pd.DataFrame({
    'PropertyName': df['PropertyName'].iloc[apartment_indices],
    'SimilarityScore': sim_scores
    })
    # Return the top 5 most similar apartments
    return recommendations
```

```python
import pandas as pd
import json

# Load the dataset
df_appartments = pd.read_csv('appartments.csv').drop(22)

# Function to parse and extract the required features from the PriceDetails column
def refined_parse_modified_v2(detail_str):
    try:
        details = json.loads(detail_str.replace("'", "\""))
    except:
        return {}

    extracted = {}
    for bhk, detail in details.items():
        # Extract building type
        extracted[f'building type_{bhk}'] = detail.get('building_type')

        # Parsing area details
        area = detail.get('area', '')
        area_parts = area.split('-')
        if len(area_parts) == 1:
            try:
                value = float(area_parts[0].replace(',', '').replace(' sq.ft.', '').strip())
                extracted[f'area low {bhk}'] = value
                extracted[f'area high {bhk}'] = value
            except:
                extracted[f'area low {bhk}'] = None
                extracted[f'area high {bhk}'] = None
        elif len(area_parts) == 2:
            try:
                extracted[f'area low {bhk}'] = float(area_parts[0].replace(',', '').replace(' sq.ft.', '').strip())
                extracted[f'area high {bhk}'] = float(area_parts[1].replace(',', '').replace(' sq.ft.', '').strip())
            except:
                extracted[f'area low {bhk}'] = None
                extracted[f'area high {bhk}'] = None

        # Parsing price details
        price_range = detail.get('price-range', '')
        price_parts = price_range.split('-')
        if len(price_parts) == 2:
            try:
                extracted[f'price low {bhk}'] = float(price_parts[0].replace('₹', '').replace(' Cr', '').replace(' L', '').strip())
                extracted[f'price high {bhk}'] = float(price_parts[1].replace('₹', '').replace(' Cr', '').replace(' L', '').strip())
                if 'L' in price_parts[0]:
                    extracted[f'price low {bhk}'] /= 100
                if 'L' in price_parts[1]:
                    extracted[f'price high {bhk}'] /= 100
            except:
                extracted[f'price low {bhk}'] = None
                extracted[f'price high {bhk}'] = None

    return extracted
# Apply the refined parsing and generate the new DataFrame structure
data_refined = []

for _, row in df_appartments.iterrows():
    features = refined_parse_modified_v2(row['PriceDetails'])
    
    # Construct a new row for the transformed dataframe
    new_row = {'PropertyName': row['PropertyName']}
    
    # Populate the new row with extracted features
    for config in ['1 BHK', '2 BHK', '3 BHK', '4 BHK', '5 BHK', '6 BHK', '1 RK', 'Land']:
        new_row[f'building type_{config}'] = features.get(f'building type_{config}')
        new_row[f'area low {config}'] = features.get(f'area low {config}')
        new_row[f'area high {config}'] = features.get(f'area high {config}')
        new_row[f'price low {config}'] = features.get(f'price low {config}')
        new_row[f'price high {config}'] = features.get(f'price high {config}')
    
    data_refined.append(new_row)

df_final_refined_v2 = pd.DataFrame(data_refined).set_index('PropertyName')
```

```python
import ast
import re
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('appartments.csv').drop(22)      
df = df.reset_index(drop=True)
```

```python
# STEP 1 — distance parser  (same logic as before, made more robust)
# ─────────────────────────────────────────────
def distance_to_meters(s):
    """Converts '1.4 Km', '450 m', '450 Meter' → float meters. Returns None on failure."""
    if not isinstance(s, str):
        return None
    s = s.strip()
    try:
        num = float(re.findall(r'[\d.]+', s)[0])
    except (IndexError, ValueError):
        return None
 
    sl = s.lower()
    if 'km' in sl:
        return num * 1000
    elif 'm' in sl:          # covers 'meter', 'meters', 'm'
        return num
    return None
```

```python
# STEP 2 — canonical name cleaner
#
# Goal: turn every raw location string into a
# lowercase, punctuation-free, whitespace-normalised
# token so that fuzzy grouping can collapse variants.
# ─────────────────────────────────────────────
def canonical(name: str) -> str:
    name = name.lower()
 
    # remove trailing/embedded city suffixes that add noise
    for suffix in [
        r'\bgurgaon\b', r'\bgurugram\b',
    ]:
        name = re.sub(suffix, '', name)
 
    # normalise punctuation & whitespace
    name = re.sub(r"['\",.\-/&()\[\]]+", ' ', name)
    name = re.sub(r'\s+', ' ', name).strip()
 
    return name
```

```python
# STEP 3 — collect ALL raw location names and
#           group them with fuzzy matching
#
# We use Python's built-in difflib (no extra install)
# to cluster names that are ≥ 0.82 similar into one
# canonical representative (the shortest clean version).
# ─────────────────────────────────────────────
from difflib import SequenceMatcher
 
def similarity(a, b):
    return SequenceMatcher(None, a, b).ratio()
 
# Gather every unique canonical name across all apartments
raw_to_canonical: dict[str, str] = {}   # raw original → canonical token
 
all_raw_names = []
for _, row in df.iterrows():
    try:
        d = ast.literal_eval(row['LocationAdvantages'])
        all_raw_names.extend(d.keys())
    except Exception:
        pass
 
unique_raws = list(dict.fromkeys(all_raw_names))   # preserves first-seen order
 
# Build canonical → list-of-raws mapping
can_to_raws: dict[str, list] = {}
for raw in unique_raws:
    c = canonical(raw)
    raw_to_canonical[raw] = c
    can_to_raws.setdefault(c, []).append(raw)
 
# Fuzzy-merge canonical names that are very similar
# Builds a mapping  canonical → merged_representative
canonical_names = list(can_to_raws.keys())
merge_map: dict[str, str] = {}          # every canonical → its cluster rep
 
THRESHOLD = 0.82
 
for i, a in enumerate(canonical_names):
    if a in merge_map:                  # already absorbed
        continue
    merge_map[a] = a                    # a is its own rep for now
    for j in range(i + 1, len(canonical_names)):
        b = canonical_names[j]
        if b in merge_map:
            continue
        if similarity(a, b) >= THRESHOLD:
            merge_map[b] = a            # b merges into a
 
# Final lookup:  raw original name → cluster representative
def final_label(raw_name: str) -> str:
    c = raw_to_canonical.get(raw_name, canonical(raw_name))
    return merge_map.get(c, c)
```

```python
# STEP 4 — build location_df
#
# Each row = one apartment.
# Each column = one merged location cluster.
# Cell value = distance in metres (minimum when
#              multiple spellings appear in the same apt).
# ─────────────────────────────────────────────
rows = []
for _, row in df.iterrows():
    try:
        raw_dict = ast.literal_eval(row['LocationAdvantages'])
    except Exception:
        raw_dict = {}
 
    entry: dict[str, float] = {}
    for loc_name, dist_str in raw_dict.items():
        col  = final_label(loc_name)
        dist = distance_to_meters(dist_str)
        if dist is None:
            continue
        # keep the closest distance when two spellings map to the same column
        if col in entry:
            entry[col] = min(entry[col], dist)
        else:
            entry[col] = dist
 
    entry['PropertyName'] = row['PropertyName']
    rows.append(entry)
 
location_df = pd.DataFrame(rows).set_index('PropertyName')
 
print(f"Shape before filtering: {location_df.shape}")
```

```python
# STEP 5 — drop noise columns
#
# Remove locations that appear in fewer than MIN_APT
# apartments — they carry almost zero signal and
# just inflate dimensionality.
# ─────────────────────────────────────────────
MIN_APT = 3          # a location must appear in at least 3 apartments
 
non_null_counts = location_df.notna().sum(axis=0)
location_df = location_df.loc[:, non_null_counts >= MIN_APT]
 
print(f"Shape after dropping rare columns (< {MIN_APT} apartments): {location_df.shape}")
```

```python
# STEP 6 — fill NaN with a large distance
#
# Missing = location not mentioned = very far away.
# We use a distance slightly beyond the maximum
# observed distance so the scaler treats it as
# "furthest possible" consistently.
# ─────────────────────────────────────────────
MAX_DIST = location_df.max().max()
FILL_VAL = MAX_DIST * 1.5           # e.g. ~1.5× the furthest real mention
location_df = location_df.fillna(FILL_VAL)
 
print(f"NaN fill value used: {FILL_VAL:.0f} metres")
```

```python
# STEP 7 — normalise and compute cosine similarity
# ─────────────────────────────────────────────
scaler = StandardScaler()
location_df_normalized = pd.DataFrame(
    scaler.fit_transform(location_df),
    columns=location_df.columns,
    index=location_df.index
)
 
cosine_sim3 = cosine_similarity(location_df_normalized)
print(f"cosine_sim3 shape: {cosine_sim3.shape}")
```

```python
# STEP 8 — quick sanity-check recommender
# ─────────────────────────────────────────────
def recommend_by_location(property_name, top_n=5):
    idx   = location_df_normalized.index.get_loc(property_name)
    scores = list(enumerate(cosine_sim3[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    top   = scores[1:top_n+1]
    return pd.DataFrame({
        'PropertyName':    location_df_normalized.index[[i for i, _ in top]].tolist(),
        'SimilarityScore': [round(s, 4) for _, s in top]
    })
 
 
print("\nTest — neighbours of 'DLF The Camellias':")
print(recommend_by_location('DLF The Camellias'))
 
print("\nTest — neighbours of 'Sobha City':")
print(recommend_by_location('Sobha City'))
```

```python
def recommend_properties_with_scores(property_name, top_n=5):
    
    cosine_sim_matrix = 0.5*cosine_sim1 + 0.8*cosine_sim2 + cosine_sim3
    # cosine_sim_matrix = cosine_sim3
    
    # Get the similarity scores for the property using its name as the index
    sim_scores = list(enumerate(cosine_sim_matrix[location_df_normalized.index.get_loc(property_name)]))
    
    # Sort properties based on the similarity scores
    sorted_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the indices and scores of the top_n most similar properties
    top_indices = [i[0] for i in sorted_scores[1:top_n+1]]
    top_scores = [i[1] for i in sorted_scores[1:top_n+1]]
    
    # Retrieve the names of the top properties using the indices
    top_properties = location_df_normalized.index[top_indices].tolist()
    
    # Create a dataframe with the results
    recommendations_df = pd.DataFrame({
        'PropertyName': top_properties,
        'SimilarityScore': top_scores
    })
    
    return recommendations_df

# Test the recommender function using a property name
recommend_properties_with_scores('DLF The Camellias')
```
