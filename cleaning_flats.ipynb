```python
def cr_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2) 
```

```python
df['price'] = df['price'].str.split(' ').apply(cr_price)
```

```python
df['price'].value_counts()
```

```python
df['price_per_sqft'].value_counts()
```

```python
#example to convert price_per_sqft to float
a = '₹ 12,250/sq.ft.'
# get text between ₹ and /sq.ft.
a = a.split('/')[0].replace('₹ ',"").replace(",","")
int(a)
```

```python
df['price_per_sqft']=df['price_per_sqft'].str.split('/').str.get(0).str.replace("₹ ","").str.replace(",","").astype(float)
#float because we have some missing values in price_per_sqft column
```
