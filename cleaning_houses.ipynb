```python
import pandas as pd
```

```python
df = pd.read_csv('houses.csv')
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.drop(['property_id','link','description','rating'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['society'].value_counts()
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','',str(name)).strip()).str.lower()
```

```python
df.dropna(subset=['price'], inplace=True)
```

```python
df['price'].value_counts()
```

```python
def clean_price(price):
    price_str = str(price).lower().strip()
    if 'lac' in price_str:
        numeric_value = float(price_str.replace('lac', '').strip())
        return numeric_value * 0.01
    return price

df['price'] = df['price'].apply(clean_price)
```

```python
df[df['price'].str.contains('lac', case=False, na=False)]
```

```python
df = df[df['price'].str.split().str.get(0) != 'Price']  # Remove header if it snuck into data
```

```python
df['price'] = df['price'].str.split().str.get(0).astype(float)
```

```python
df['price'].value_counts()# now the price is in crores
```

```python
df.head(5)
```

```python
df['rate'] = df['rate'].str.split('/').str.get(0).str.replace(',','').str.replace('₹','').astype(float)
```

```python
df['rate'].value_counts()
```

```python
df.rename(columns = {'rate':'price_per_sqft'}, inplace=True)
```

```python
#area
import re
df['areaWithType'].value_counts()
```

```python
def get_area(s):
    if s.split(' ')[2] == 'area:':
        return s.split(' ')[3]
    else:
        return s.split(' ')[2].split('(')[0]
```

```python
df[df['areaWithType'].str.split(' ').str.get(2)!= 'area:']
```

```python
df['area'] = df['areaWithType'].apply(get_area)
```

```python
df[df['area']=='60Built']
```

```python
df['area']=df['area'].str.replace('Built','')
```

```python
df['area'] = df['area'].astype(float)
```

```python
df.head(3)
```

```python
df['bedRoom']=df['bedRoom'].str.split(' ').str.get(0).astype(int)
```

```python
df['balcony'] = df['balcony'].str.split(" ").str.get(0)
```

```python
df['bathroom'] = df['bathroom'].str.split(" ").str.get(0).astype(int)
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df['noOfFloor'] = df['noOfFloor'].str.split(' ').str.get(0)
```

```python
df['facing'].fillna('NA', inplace=True)
```

```python
df.insert(loc=2,column='property_type',value = 'house')
```

```python
df.insert(loc=5,column='sector',value = df['property_name'].str.split("in").str.get(1).str.replace('Gurgaon','').str.replace('Faridabad','').str.strip())
```
